6.  Explain the concept of generators in Python and how they are defined.
    - Generators in Python are special types of iterators that allow you to generate values on the fly instead of storing them in memory all at once. They are useful for working with large datasets or infinite sequences because they produce one value at a time as needed.

        Generators are defined using:               

        The yield keyword (instead of return).

        Generator expressions (similar to list comprehensions).

        Defining Generators

        Using yield in a Function: A generator function is defined like a normal function, but it uses yield instead of return to produce values one at a time.

        Example:

        def my_generator():            
        yield 1              
        yield 2            
        yield 3
        gen = my_generator()  # Creates a generator object          
        print(next(gen))  # Output: 1             
        print(next(gen))  # Output: 2            
        print(next(gen))  # Output: 3        
         #print(next(gen))  # Raises StopIteration
        
        Using Generator Expressions

        Generators can also be created using a compact syntax similar to list comprehensions, but with parentheses (  ) instead of square brackets [  ].

        Example:

        gen = (x**2 for x in range(5))  # Generator expression

         print(next(gen))  # Output: 0     
        print(next(gen))  # Output: 1       
        print(next(gen))  # Output: 4


7. What are the advantages of using generators over regular functions?
    -  Generators offer several advantages over regular functions. They are:

       Memory Efficiency: Generators produce items one at a time and only when needed, which is ideal for working with large datasets or infinite sequences without consuming a lot of memory.

        Lazy Evaluation: They generate values on the fly, which means they only compute values as they are requested, leading to better performance in many cases.

        Simplified Code: Using yield in a generator function can make the code simpler and more readable compared to using complex loops and data structures.

        State Retention: Generators maintain their state between calls, allowing them to resume where they left off, which can be useful for tasks like reading large files or streaming data.

        Improved Performance: Since generators yield values one at a time, they can start producing results immediately without waiting to generate the entire sequence, leading to faster initial response times.

        These benefits make generators a powerful tool for efficient and effective data.

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - The map(), reduce(), and filter() functions in Python are all used for functional programming, but they serve different purposes:

       ## map( ):                  

      Purpose:   Applies a given function to each item in an iterable (like a list) and returns a map object (an iterator) with the results.

       Usage: Useful for transforming data.

       Example: map(lambda x: x * 2, [1, 2, 3]) results in [2, 4, 6].

       ## reduce( ):

       Purpose: Applies a given function cumulatively to the items of an iterable, reducing the iterable to a single value.

       Usage: Useful for performing cumulative operations like summing or multiplying all elements.

       Example: reduce(lambda x, y: x + y, [1, 2, 3, 4]) results in 10.

       Note: reduce() is not a built-in function in Python 3; it is available in the functools module.

       ## filter( ):

       Purpose: Applies a given function to each item in an iterable and returns an iterator with only the items for which the function returns True.

       Usage: Useful for filtering data.

       Example: filter(lambda x: x % 2 == 0, [1, 2, 3, 4]) results in [2, 4].

       These functions are powerful tools for data processing and manipulation in Python, each serving a unique role in functional programming.

the list.

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
# Ans

class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.n <= 0:
            raise StopIteration
        current = self.a
        self.a, self.b = self.b, self.a + self.b
        self.n -= 1
        return current
fibonacci_sequence = FibonacciIterator(10)
for num in fibonacci_sequence:
    print(num)
```

6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
# Ans

def powers_of_2_yield(n):
    for i in range(n + 1):
        yield 2 ** i
for power in powers_of_2_yield(5):
    print(power)
```

7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
# Ans

file_content = [
    "Hello, this is the first line.",
    "This is the second line.",
    "And this is the third line.",
    "This is the fourth line.",
]
def read_file_line_by_line(content):
    for line in content:
        yield line.strip()
for line in read_file_line_by_line(file_content):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# Ans

celsius_temperatures = [20,40,22,25]
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print(fahrenheit_temperatures)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Ans

def is_not_vowel(char):
    return char.lower() not in 'aeiou'
input_string = "Hello, World!"
filtered_string = ''.join(filter(is_not_vowel, input_string))
print(filtered_string)
```

11.  Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:                             
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.                              
Write a Python program using lambda and map.

```python
# Ans

orders = [
    [101, 4, 40.95],
    [202, 5, 56.80],
    [302, 3, 32.95],
    [402, 3, 24.99],
]
def calculate_order(order):
    order_number, price_per_item, quantity = order
    total_price = price_per_item * quantity
    if total_price < 100:
        total_price += 10
    return (order_number, total_price)
result = list(map(lambda order: calculate_order(order), orders))
print(result)
```
